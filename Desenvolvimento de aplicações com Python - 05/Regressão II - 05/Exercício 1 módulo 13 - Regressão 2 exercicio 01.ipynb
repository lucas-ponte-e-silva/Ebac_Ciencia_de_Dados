1. Ajuste um modelo para prever log(renda) considerando todas as covariáveis disponíveis.

2. Remova a variável menos significante e analise:

3. Siga removendo as variáveis menos significantes, sempre que o *p-value* for menor que 5%. Compare o modelo final com o inicial. Observe os indicadores e conclua se o modelo parece melhor. 
    

```python
# 01 - Ajuste um modelo para prever log(renda) considerando todas as covariáveis disponíveis.

import pandas as pd
import numpy as np
import statsmodels.api as sm
from patsy import dmatrices

y, X = dmatrices('log_renda ~ C(sexo, Treatment ("F")) + C(posse_de_veiculo, Treatment (False)) + C(posse_de_imovel, Treatment (True)) + qtd_filhos + C(tipo_renda, Treatment ("Assalariado")) + C(educacao, Treatment ("Secundário")) + C(estado_civil, Treatment ("Solteiro")) + C(tipo_residencia, Treatment ("Casa")) + idade + tempo_emprego + qt_pessoas_residencia + mau + renda' , data=df, return_type='dataframe')

modelo_log_renda = sm.OLS(y, X).fit().summary()
modelo_log_renda
```

```python
#02) Remova a variável menos significante e analise:
df = df[df['estado_civil'] != 'Solteiro']

print(" Removi a variavel 'estado_civil Solteiro' , pois é a variavel menos significante da tabela")
```

```python
y2, X2 = dmatrices('log_renda ~ C(sexo, Treatment ("F")) + C(posse_de_veiculo, Treatment (False)) + C(posse_de_imovel, Treatment (True)) + qtd_filhos + C(tipo_renda, Treatment ("Assalariado")) + C(educacao, Treatment ("Secundário")) + C(estado_civil) + C(tipo_residencia, Treatment ("Casa")) + idade + tempo_emprego + qt_pessoas_residencia + mau + renda' , data=df, return_type='dataframe')

modelo_log_renda_2 = sm.OLS(y2, X2).fit().summary()
modelo_log_renda_2
```

```python
#03) Siga removendo as variáveis menos significantes, sempre que o p-value for menor que 5%. Compare o modelo final com o inicial. Observe os indicadores e conclua se o modelo parece melhor.
df = df[df['tipo_renda'] != 'Bolsista']
df = df[df['tipo_renda'] != 'Pensionista']
df = df[df['educacao'] != 'Primário']
df = df[df['educacao'] != 'Superior incompleto']
df = df[df['estado_civil'] != 'Separado']
df = df[df['estado_civil'] != 'União']
df = df[df['estado_civil'] != 'Viúvo']
df = df[df['tipo_residencia'] != 'Comunitário']
df = df[df['tipo_residencia'] != 'Estúdio']
df = df[df['tipo_residencia'] != 'Governamental']
df = df[df['mau'] != 'True']

print(" Removi as variáveis com p-valor maior que 0.05 ")
```

```python

y3, X3 = dmatrices('log_renda ~ C(sexo, Treatment ("F")) + C(posse_de_veiculo, Treatment (False)) + C(posse_de_imovel, Treatment (True)) + C(tipo_renda, Treatment ("Assalariado")) + C(educacao, Treatment ("Secundário")) + C(estado_civil) + C(tipo_residencia, Treatment ("Casa")) + idade + tempo_emprego + mau + renda' , data=df, return_type='dataframe') 
                   
modelo_log_renda_3 = sm.OLS(y3, X3).fit().summary()
modelo_log_renda_3
```
